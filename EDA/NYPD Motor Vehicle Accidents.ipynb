Persons/PEDESTRIANS/CYCLIST/ - INJURED/KILLED  by month

```python
inj_kill = df_inj_kill.DATENew.dt.month
s = df_inj_kill.groupby(inj_kill)
dfinj_kill = s.sum()
month =dfinj_kill.index

dfinj =dfinj_kill[['NUMBER OF PERSONS INJURED','NUMBER OF PEDESTRIANS INJURED','NUMBER OF CYCLIST INJURED','NUMBER OF MOTORIST INJURED']]
dfkill =dfinj_kill[['NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST KILLED']]
```

```python
dfinj.plot(x=month, figsize=(12,6), kind = 'bar', title = ' Injured',width=1)
plt.ylabel('inj')
plt.xlabel('month')
plt.show()
```

```python
dfkill.plot(x=month, figsize=(12,6), kind = 'bar', title = 'Killed',width=1)
plt.ylabel('kill')
plt.xlabel('month')
plt.show()
```

Persons/PEDESTRIANS/CYCLIST/ - INJURED/KILLED by Year

```python
inj_kill = df_inj_kill.DATENew.dt.year
s = df_inj_kill.groupby(inj_kill)
dfinj_kill_year = s.sum()
year =dfinj_kill_year.index

dfinj_year =dfinj_kill_year[['NUMBER OF PERSONS INJURED','NUMBER OF PEDESTRIANS INJURED','NUMBER OF CYCLIST INJURED','NUMBER OF MOTORIST INJURED']]
dfkill_year =dfinj_kill_year[['NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS KILLED','NUMBER OF CYCLIST KILLED','NUMBER OF MOTORIST KILLED']]
```

```python
dfinj_year.plot(x=year, figsize=(12,6), kind = 'bar', title = 'injured per year',width=1)
plt.ylabel('inj_year')
plt.xlabel('year')
plt.show()
```

```python
dfkill_year.plot(x=year, figsize=(12,6), kind = 'bar', title = 'Killed',width=1)
plt.ylabel('kill_year')
plt.xlabel('year')
plt.show()
```
